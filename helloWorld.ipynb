```python
from qiskit import *

qr = QuantumRegister(2)
cr = ClassicalRegister(2)
myCircuit = QuantumCircuit(qr, cr) #initiate quantum circuit 
%matplotlib inline

myCircuit.draw()
```

```python
myCircuit.h(qr[0]) #hademar gate at qr[0]
```

```python
myCircuit.draw(output="mpl") #output formatted to matplotlib
```

```python
myCircuit.cx(qr[0], qr[1]) #0 is the control register, 1 is the target register
```

```python
myCircuit.draw(output="mpl")
```

```python
myCircuit.measure(qr, cr)
```

```python
myCircuit.draw(output = "mpl")
```

```python
from qiskit import Aer
from qiskit.tools.visualization import plot_histogram
simulator = Aer.get_backend("qasm_simulator")
result = execute(myCircuit, backend = simulator).result()
plot_histogram(result.get_counts(myCircuit))
```
